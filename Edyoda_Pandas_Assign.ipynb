```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/covid_data.csv')
df.head()
```

```python
# convert abpve to comprehensive list commmand
cat_col = [x for x in df.dtypes.index if df.dtypes[x] == 'object']
print(cat_col)
```

```python
# a. Find count of unique values in location column.
print("Unique values in location column:", df['location'].nunique())

# b. Find which continent has maximum frequency using values counts.
print("\nContinent with maximum frequency:\n", df['continent'].value_counts().head(1))

# c. Find maximum & mean value in 'total_cases'.
print("\nMaximum total_cases:", df['total_cases'].max())
print("Mean total_cases:", df['total_cases'].mean())

# d. Find 25%,50% & 75% quartile value in 'total_deaths'.
print("\nQuartiles for total_deaths:\n", df['total_deaths'].quantile([0.25, 0.5, 0.75]))

# e. Find which continent has maximum 'human_development_index'.
max_hdi_continent = df.loc[df['human_development_index'].idxmax(), 'continent']
print("\nContinent with maximum human_development_index:", max_hdi_continent)

# f. Find which continent has minimum 'gdp_per_capita'.
min_gdp_continent = df.loc[df['gdp_per_capita'].idxmin(), 'continent']
print("\nContinent with minimum gdp_per_capita:", min_gdp_continent)
```

```python
# a. Remove all duplicates observations
df = df.drop_duplicates()

# b. Find missing values in all columns
print("Missing values before cleaning:\n", df.isnull().sum())

# c. Remove all observations where continent column value is missing
df.dropna(subset=['continent'], inplace=True)

# d. Fill all missing values with 0
df.fillna(0, inplace=True)

print("\nMissing values after cleaning:\n", df.isnull().sum())
display(df.head())
```

```python
# a. Convert date column in datetime format using pandas.to_datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')

# b. Create new column month after extracting month data from date column.
df['month'] = df['date'].dt.month

display(df.head())
```

```python
# a. Find max value in all columns using groupby function on 'continent' column
df_groupby = df.groupby('continent').max().reset_index()

# b. Store the result in a new dataframe named 'df_groupby'.
# This was done in the previous line as part of the reset_index() call

display(df_groupby)
```

```python
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']

display(df_groupby.head())
```

```python
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
sns.histplot(df_groupby['gdp_per_capita'], kde=True)
plt.title('Distribution of GDP per Capita by Continent')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(data=df_groupby, x='gdp_per_capita', y='total_cases')
plt.title('Total Cases vs. GDP per Capita by Continent')
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
plt.show()
```

```python
# c. Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
plt.show()
```

```python
# d. Plot a bar plot of 'continent' column with 'total_cases'
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar', aspect=2)
plt.title('Total Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
# Find the total number of cases till date for each country
total_cases_per_country = df.groupby('location')[['total_cases', 'total_deaths']].max().reset_index()
print("Total cases till date for each country:")
display(total_cases_per_country.head())

# Find the country with the maximum number of cases
country_max_cases = total_cases_per_country.loc[total_cases_per_country['total_cases'].idxmax()]
print("\nCountry with maximum total cases:")
display(country_max_cases)

# Find the country with the least number of cases
country_min_cases = total_cases_per_country.loc[total_cases_per_country['total_cases'].idxmin()]
print("\nCountry with least total cases:")
display(country_min_cases)

# Calculate the death-to-case ratio for each country
# We need to handle potential division by zero if total_cases is 0
total_cases_per_country['death_to_case_ratio'] = total_cases_per_country['total_deaths'] / total_cases_per_country['total_cases']
total_cases_per_country['death_to_case_ratio'] = total_cases_per_country['death_to_case_ratio'].fillna(0) # Fill NaN with 0 for countries with 0 cases

# Find the country with the maximum ratio of death to cases
country_max_ratio = total_cases_per_country.loc[total_cases_per_country['death_to_case_ratio'].idxmax()]
print("\nCountry with maximum death to case ratio:")
display(country_max_ratio)
```
